```python
import numpy as np
import matplotlib.pyplot as plt
```

# Example
Let's consider the function : $$f(x) = 7e^{0.5x}$$
The derivative of this function is : $$f'(x) = 7\times 0.5e^{0.5x} = 3.5e^{0.5x}$$

At $x = 2$, the derivative of the function is $$f'(2) = 3.5e^{0.5 \times 2} = 3.5e = 9.513986399606658$$

Let's approximate the derivative of the function at x = 2 using the definition of the derivative for different values of h using FDD, BDD and CDD.

```python
def f(x):
    return 7 * np.exp(0.5 * x)

def f_prime(x):
    return 3.5 * np.exp(0.5 * x)
```

```python
true_value = 9.513986399606658
```

### Recall : Absolute Relative True Error

$$ Absolute Relative True Error = | \frac{True Value - Approximate Value}{True Value} | \times 100\% $$

# Forward Divided Difference (FDD)
The forward difference derivative is defined as:
$$
f'(x) = \frac{f(x+h)-f(x)}{h}
$$
where $h$ is a small number. The smaller $h$ is, the more accurate the derivative is. However, if $h$ is too small, the derivative will be inaccurate due to round-off error.

```python
def fdd(f, x, h):
    return (f(x+h) - f(x)) / h
```

```python
# let's find the derivatives of f(x) at x = 2 for different values of h
# and find the errors
# store the errors in a list for future plotting

h = [0.3, 0.15, 0.1, 0.05, 0.01, 0.001]
x = 2
fdd_errors = []
for i in h:
    approx_value = fdd(f, x, i)
    arte = (abs(true_value - approx_value)) * 100 /true_value
    fdd_errors.append(arte)
    print("h = ", i, "; derivative = ", approx_value, "; ARTE = ", arte, "%")
```

# Backward Divided Difference (BDD)
The backward divided difference is defined as:
$$
\frac{df}{dx} = \frac{f(x) - f(x - h)}{h}
$$
where $h$ is a small number. The smaller $h$ is, the more accurate the derivative is. However, if $h$ is too small, the derivative will be inaccurate due to round-off error.

```python
def bdd(f, x, h):
    return (f(x) - f(x-h)) / h
```

```python
# let's find the derivatives of f(x) at x = 2 for different values of h
# and find the errors
# store the errors in a list for future plotting

true_value = 9.513986399606658
h = [0.3, 0.15, 0.1, 0.05, 0.01, 0.001]
x = 2
bdd_errors = []
for i in h:
    approx_value = bdd(f, x, i)
    arte = (abs(true_value - approx_value)) * 100 /true_value
    bdd_errors.append(arte)
    print("h = ", i, "; derivative = ", approx_value, "; ARTE = ", arte, "%")
```

# Central Divided Difference (CDD)
The central difference derivative is defined as:
$$
\frac{df}{dx} = \frac{f(x + h) - f(x - h)}{2h}
$$
where $h$ is a small number. The smaller $h$ is, the more accurate the derivative is. However, if $h$ is too small, the derivative will be inaccurate due to round-off error.

```python
def cdd(f, x, h):
    return (f(x + h) - f(x-h)) / (2*h)
```

```python
# let's find the derivatives of f(x) at x = 2 for different values of h
# and find the errors
# store the errors in a list for future plotting

true_value = 9.513986399606658
h = [0.3, 0.15, 0.1, 0.05, 0.01, 0.001]
x = 2
cdd_errors = []
for i in h:
    approx_value = cdd(f, x, i)
    arte = (abs(true_value - approx_value)) * 100 /true_value
    cdd_errors.append(arte)
    print("h = ", i, "; derivative = ", approx_value, "; ARTE = ", arte, "%")
```

# Plotting

Let's plot the FDD, BDD and CDD errors with respect to step size in the same plot.

```python
plt.figure(figsize=(10, 6))
plt.plot(h, fdd_errors, label='FDD', marker='o')
plt.plot(h, bdd_errors, label='BDD', marker='^')
plt.plot(h, cdd_errors, label='CDD', marker='x')

plt.xlabel('h', fontsize=14)
plt.ylabel('$ |\epsilon_t| $ %', fontsize=14)
plt.title('Error vs. Step Size for Differentiation Methods', fontsize=16)
plt.show()
```

We can see from the graph that the bigger the step size, the bigger the approximation error. From the graph above, it's evident that both FDD and BDD have similar error levels, whereas CDD exhibits significantly lower errors for the same step size.

# Another Example
Let's consider the following function :
$$ f(x) = x^3 - 3x^2 + 5x - 6 $$
The derivative of the function is :
$$ f'(x) = 3x^2 - 6x + 5 $$

Let's take 1000 points in the interval [-1, 1] and find the approximated derivative of the function for these values using FDD, BDD and CDD. Let's also find the true derivative values in those points. We are going to plot these approximations and the true value in a graph.

```python
def new_func(x):
    return x**3 - 3*x**2 + 5*x - 6

def derivative_of_new_func(x):
    return 3*x**2 - 6*x + 5
```

```python
x = np.linspace(-1, 1, 1000)
h = 0.5             # play with this value to see how it affects the results

der_fdd = fdd(new_func, x, h)
der_bdd = bdd(new_func, x, h)
der_cdd = cdd(new_func, x, h)
der_true = derivative_of_new_func(x)

plt.figure(figsize=(10, 5))
plt.plot(x, der_fdd, label='FDD')
plt.plot(x, der_bdd, label='BDD')
plt.plot(x, der_cdd, label='CDD')
plt.plot(x, der_true, label='True')
plt.xlabel('$x$', fontsize=15)
plt.ylabel('$f\'(x)$', fontsize=15)
plt.title('Comparison of Differentiation Methods', fontsize=20)
plt.legend()
plt.show()
```

The CDD curve closely aligns with the True curve, showing the smallest gap. In contrast, the BDD and FDD curves have a larger gap compared to the True curve. This gap between the True curve and the Approximation curves serves as an indicator of the errors.
